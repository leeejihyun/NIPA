### The first 20 classes by person copy & paste

```python
import pandas as pd
```

```python
plant_label_list = []
disease_label_list = []

with open('./trial.tsv', 'r') as f:
    while True:
        line = f.readline()
        file_name = line.rstrip().split('\t')[0]
        if file_name == '20.jpg':
            break
        plant_label = line.rstrip().split('\t')[1]
        plant_label_list.append(plant_label)
        disease_label = line.rstrip().split('\t')[2]
        disease_label_list.append(disease_label)

plant_label_list, disease_label_list
```

```python
file_name_list = [str(index) + '.jpg' for index in range(4000)]
file_name_list[:3], file_name_list[-1], len(file_name_list)
```

```python
plant_label_list *= 200
plant_label_list[20:40], plant_label_list[-1]
```

```python
disease_label_list *= 200
disease_label_list[20:40], disease_label_list[-1]
```

```python
len(plant_label_list)
```

```python
import pandas as pd

sub = pd.DataFrame({'file_name': file_name_list,
             'plant_label': plant_label_list,
             'disease_label': disease_label_list})
sub.head(20)
```

```python
sub.to_csv('sub_person.tsv', index=False, header=None, sep='\t')
```
