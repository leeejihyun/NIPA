```python
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model
import numpy as np
import os
from PIL import Image
```

```python
path_dir = 'test'
file_list = os.listdir(path_dir)
file_list = [int(file.split('.')[0]) for file in file_list if file.endswith(".jpg")] 
file_list[:5], len(file_list)
```

```python
file_list.sort()
file_list[:5]
```

```python
file_list = [str(file) + '.jpg' for file in file_list]
file_list
```

```python
# load and prepare the image
def load_image(filename):
    # load the image
    img = load_img(filename) 
    # convert to array
    img = img_to_array(img)
    # reshape into a single sample with 1 channel
    img = img.reshape(1, 256, 256, 3)
    # prepare pixel data
    img = img.astype('float32')
    img = img / 255.0
    return img

# load an image and predict the class
def run_example():
    # load the image
    img = load_image('test/0.jpg')
    # load model
    model = load_model('b_val_acc_0.90.h5')
    # predict the class
    pred = model.predict(img)
    pred_label = np.argmax(pred[0])
    
    print(pred[0])
    print('예측 레이블:', pred_label)
    
# entry point, run the example
run_example() 
```

```python
y_pred_test = []
count = 0
for i in file_list:
    img = load_image("test/{0}".format(i))
    model = load_model('b_val_acc_0.90.h5')
    pred = model.predict(img)
    pred_label = np.argmax(pred[0])
    y_pred_test.append(pred_label)
    count += 1
    print(count)
    if count == 1000:
        break
print(y_pred_test)
```
